```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy as sc

df=pd.read_csv('train.csv')
df_test=pd.read_csv('test.csv')
df
```

```python
#df['MSZoning']=df['MSZoning'].astype('category').cat.codes

for x in [df.columns,df_test.columns]:
    for i in x:
        df[i]=df[i].astype('category').cat.codes #Kategoorik verilere dönüşüm

        if df[i].isna().sum()!=0:
            df[i].fillna(df[i].mode(),inplace=True) #Eksik veri varsa moduna göre hareket ediliyor
        else:
            if df[i].isna().sum()!=0:
                df[i].fillna(df[i].mean(),inplace=True) #Eksik verilere ortalama alınarak yaklaşılıyor
```

## 4. MODEL

```python
X_train=df.drop("SalePrice",axis=1)
y_train=df["SalePrice"]

X_test=df.drop("SalePrice",axis=1)
y_test=df["SalePrice"]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

#### 4.1 Hyperparameter

```python
from sklearn.model_selection import cross_validate, GridSearchCV
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVC
from sklearn.neural_network import MLPRegressor
from sklearn.linear_model import SGDRegressor
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.metrics import classification_report, accuracy_score, f1_score, confusion_matrix


regressor = {
    "NB": GaussianNB(),
    "SGD": SGDRegressor(),
    "KNN": KNeighborsRegressor(), 
    "LR": LogisticRegression(max_iter=1000), 
    "DT": DecisionTreeRegressor(),
    "RF": RandomForestRegressor(n_estimators=100, max_depth=3, random_state=2),
    "SVM": SVC(),
    "MLP": MLPRegressor(max_iter=1000),
    "XGB": XGBRegressor(),
    "LGBM": LGBMRegressor()
}

results = pd.DataFrame(columns=["Regressor", "Avg_Accuracy", "Avg_F1_Score"])
for name, clf in regressor.items():
    model = clf
    cv_results = cross_validate(
        model, X_train_scaled, y_train, cv=10,
        scoring=(['accuracy', 'f1'])
    )

    results = results.append({
        "Regressor": name,
        "Avg_Accuracy": cv_results['test_accuracy'].mean(),
        "Avg_F1_Score": cv_results['test_f1'].mean()
    }, ignore_index=True)
    
results["Avg_Overall"] = (results["Avg_Accuracy"] + results["Avg_F1_Score"]) / 2
results = results.sort_values("Avg_Overall", ascending=False)
results
```

```python
import seaborn as sbn
plt.figure(figsize=(12, 6))
sbn.barplot(data=results, x="Avg_Overall", y="Classifier")
plt.title("Average Overall CV Score")
plt.show()
```

#### 4.2 Normal Çözüm

```python
from sklearn.model_selection import cross_validate, GridSearchCV
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVC
from sklearn.neural_network import MLPRegressor
from sklearn.linear_model import SGDRegressor
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.metrics import classification_report, accuracy_score, f1_score, confusion_matrix
```

```python
regressor = RandomForestRegressor()
params = {
    'max_features':(0.5,1),
    'min_samples_split':(2,10,20),
    'max_depth': (None,1,2,3,4),
    'n_estimators': (50,100,150,200,250,300),
}
clf = GridSearchCV(regressor, params, cv=10)
clf.fit(X_train_scaled, y_train)
print("Best hyperparameter:", clf.best_params_)
```

```python
predictions = clf.predict(X_test)


rmse = np.sqrt(np.square(y_test - predictions)).mean()


r2 = 1 - np.square(y_test - predictions).sum()/np.square(y_test - y_test.mean()).sum()

print(f"testing data: root mean squared error = {rmse}\nR-squared = {r2}")
```

```python
predictions_train = clf.predict(X_train)

rmse_train = np.sqrt(np.square(y_train - predictions_train)).mean()

r2_train = 1 - np.square(y_train - predictions_train).sum()/np.square(y_train - y_train.mean()).sum()

print(f"training data: root mean squared error = {rmse_train}\nR-squared = {r2_train}")
```
